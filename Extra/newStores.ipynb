```python
import urllib
from bs4 import BeautifulSoup
import pandas as pd
stores=pd.read_csv("../csv/storeDetails/orginalScrapped.csv")
storelink=stores['Link']
storeName=stores['name']
storeRanking=stores['Ranking']
name=[]
ranking=[]
link=[]
apps=[]
AppNames=['pushcrew',
          'wistia',
          'powr',
          'optimonk',
          'optimizely',
          'loox',
          'yotpo',
          'judge',
          'product',
          'jozi',
          'feefo',
          'globo',
           'recart',
           'kit',
           'manycart',
          'salespop',
         'shopify',
         'chatchamp',
         'appsbysimple']
for i in range(0,5):
    slink=storelink[i]
    appList=[]
    r = urllib.urlopen(slink).read()
    soup = BeautifulSoup(r,"lxml")
    scriptTags=soup.find_all("script")
    script=str(scriptTags)
    print(slink)
    for app in AppNames:
        if (app in script):
            appList.append(app)
    name.append(storeName[i])
    ranking.append(storeRanking[i])
    link.append(slink)
    apps.append(appList)
    print"=============================="
newStores=pd.DataFrame({"name":name,"link":link,"ranking":ranking,"apps":apps})
newStores
```
